```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data for the normal distribution
mu = 0  # mean
sigma = 1  # standard deviation
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = (1/(sigma * np.sqrt(2*np.pi))) * np.exp(-0.5*((x-mu)/sigma)**2)

# Get the chi square
chi_square = np.sum(y)
print(chi_square)


# Plot the normal distribution
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('Probability density')
plt.title('Normal Distribution')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate normal data
np.random.seed(42)
data = np.random.normal(loc=0, scale=1, size=1000)

# Rank the data
ranks = stats.rankdata(data)

# Convert ranks to normal scores (rankits)
normal_scores = stats.norm.ppf((ranks - 0.5) / len(ranks))

# Plot the original data and the normal scores
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(data, bins=50, density=True, alpha=0.6, color='b')
plt.title('Histogram of Normal Data')

plt.subplot(1, 2, 2)
plt.hist(normal_scores, bins=50, density=True, alpha=0.6, color='r')
plt.title('Histogram of Normal Scores')

plt.tight_layout()
plt.show()
```

```python
# Plot the normal distribution variable squared
plt.plot(x, y**2)
plt.xlabel('x')
plt.ylabel('Probability density')
plt.title('Normal Distribution Squared')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Degrees of freedom
df = 1

# Generate data for the chi-square distribution
x = np.sqrt(np.linspace(1, 30, 100))
y = stats.chi2.pdf(x, df)

# Generate Zipf data
a = 2
x_zipf = np.arange(1, 30)
y_zipf = stats.zipf.pmf(x_zipf, a)

# Plot the chi-square distribution
plt.plot(x, y)
plt.xlabel('x')

plt.plot(x_zipf, y_zipf, 'ro')

plt.ylabel('Probability density')
plt.title('Chi-Square Distribution')
plt.grid(True)
plt.show()
```
